#### Ejercicio 3

Utiliza esta formula para escribir una funcion llamada  ́ binomial(n,k) (o como
tu quieras) que calcule el coeficiente binomial para un n y k dados. Asegurate de  ́
que tu funcion devuelva la respuesta en forma de un n  ́ umero entero (no flotante)  ́
y proporcione el valor correcto de 1 para el caso en que k = 0.

```python
#Definimos la función binomial

from math import factorial

n=5 
k=4

def binomial(n,k):
    if k==0:
        bin=1
    elif n<k:
        print("k no puede ser mayor a n")
    else:
        bin=((factorial(n))/(factorial(k)*factorial(n-k)))

    return bin

print("El coeficiente binomial (n,k) es ", int(binomial(n,k)))
```

b) Usando tu función, escribe un programa que imprima las primeras 20 líneas del "triángulo de Pascal".

La n-ésima línea del triángulo de Pascal contiene n + 1 números, que son los coeficientes

$$
 \binom{n}{0}, \binom{n}{1}
$$
 , y así sucesivamente hasta $$ \binom{n}{n} $$. De tal manera que las primeras líneas son:



\begin{array}{l}
1 \\
1 \quad 1 \\
1 \quad 2 \quad 1 \\
1 \quad 3 \quad 3 \quad 1 \\
1 \quad 4 \quad 6 \quad 4 \quad 1
\end{array}


```python
for n in range(1, 21): # ciclo para producir el triángulo de pascal
    linea_n = [1]
    for k in range(1, n):
        linea_n.append(int(binomial(n-1, k)))
    linea_n.append(1)
    print("Línea ", n, ":", linea_n)
```

c) Escribe un programa para calcular:
1) la probabilidad total de que una moneda lanzada 100 veces, salga aguila exactamente 60 veces y
2) la probabilidad de que salga aguila 60 veces o más.

```python
def probabilidad(n,k): #función para calcular la probabilidad de una moneda no sesgada
    probabilidad = binomial(n,k) / (2**n)
    return probabilidad

#1) Para que salga águila 60 veces

print("La probabilidad de que tengamos exactamente 60 águilas es de: ", probabilidad(100,60))

#2) Para que salga águila 60 veces o más

probabilidad_total = 0
for k in range(60,101):
    probabilidad_total += probabilidad(100,k)

print("La probabilidad de que tengamos 60 águilas o más es de: ", probabilidad_total)
```

```python
def catalan(n):
    if n==0:
        return 1
    else:
        return ((4*n-2)/(n+1))*catalan(n-1)
    
print("El número de catalan para n=100 es: ", int(catalan(100)))
```

### b) Euclides demostró que el máximo común divisor $$ g(m, n) $$ de dos enteros no negativos m y n satisface que:

$$

g(m, n) = 
\begin{cases} 
m & \text{si } n = 0, \\
g(n, m \mod n) & \text{si } n > 0.
\end{cases}
$$



Usando esta fórmula, escribe una función `g(m, n)` que emplee recursividad para calcular el máximo común divisor de m y n. Usa tu función para calcular e imprimir el máximo común divisor de 108 y 192.


```python
def MCD(m,n):
    if n==0:
        return m
    else:
        return MCD(n,m%n)
    
print("El máximo común divisor de 108 y 192 es: ", MCD(108,192))
```
