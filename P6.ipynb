```python
import random
#para más orden, podremos todo lo anterior en funciones separadas

def escribir_matriz(matriz,archivo):
    with open(archivo,"w") as f:
        for fila in matriz:
            f.write(" ".join(map(str,fila))+"\n") #agregamos map porque de otra forma arrojaba error

def leer_matriz(archivo):
    matriz=[]
    with open(archivo,"r") as f:
        for linea in f:
            fila=[float(x) for x in linea.strip().split()]
            matriz.append(fila)
    return matriz

def producto_generalizado(a,b):
    n=len(a)
    m=len(a[0]) #columnas de a
    p=len(b[0]) #columnas de b

    if m!=len(b): #verificamos que podamos hacer el producto
        print("Las dimensiones de las matrices no son compatibles para el producto")
        return None
    else:
        resultado=[[0 for _ in range(p)] for _ in range(n)] #inicializamos

    for i in range(n):
        for j in range(p):
            suma=0
            for k in range(m):
                suma+=a[i][k]*b[k][j]
            resultado[i][j]=suma
    return resultado

#esta función será la que llamaremos para realizar el producto, es la que interactua con el usuario
def procesar_matrices():
    #le pediremos todo al usuario en cuanto a dimensiones y verificamos que son datos útiles
    n=int(input("Ingrese el número de filas de la matriz A (n): "))
    while n<=0:
        print("El número de filas debe ser mayor a 0.")
        n=int(input("Ingrese el número de filas de la matriz A (n): "))

    m=int(input("Ingrese el número de columnas de la matriz A (m): "))
    while m<=0:
        print("El número de columnas debe ser mayor a 0.")
        m=int(input("Ingrese el número de columnas de la matriz A (m): "))

    p=int(input("Ingrese el número de columnas de la matriz B (p): "))
    while p<=0:
        print("El número de columnas debe ser mayor a 0.")
        p=int(input("Ingrese el número de columnas de la matriz B (p): "))

    #generamos las matrices aleatorias con las dimensiones dadas
    matriz_a=[[random.random()*100 for _ in range(m)] for _ in range(n)]
    matriz_b=[[random.random()*100 for _ in range(p)] for _ in range(m)]

    #escribimos las matrices en archivos
    escribir_matriz(matriz_a,"matriz1.txt")
    escribir_matriz(matriz_b,"matriz2.txt")

    #leemos los archivos
    matriz1=leer_matriz("matriz1.txt")
    matriz2=leer_matriz("matriz2.txt")

    #calculamos el producto
    p=producto_generalizado(matriz1,matriz2)
    if p is not None:
        print("El producto de las matrices:")
        print("Matriz A:")
        for fila in matriz1:
            print(fila)
        print("\nMatriz B:")
        for fila in matriz2:
            print(fila)
        print("\nResultado:")
        for fila in p:
            print(fila)
    else:
        print("No se puede calcular el producto de las matrices")

procesar_matrices()
```

a) A partir del programa para la Eliminacion Gaussiana que hicimos en clase, escribe una programa que calcule la descomposicion LU de una matriz. De hecho, el calculo es el mismo que para la  ́ eliminaci ́on gaussiana, excepto que en cada paso del calculo es debes extraer los elementos apropiados de la matriz y ensamblarlos para formar la matriz diagonal inferior L

```python
import numpy as np

#sabemos que la descomposición LU consiste en separar una matriz
#en el producto de una matriz triangular inferior L y una matriz triangular superior U

#definimos la función como en clase
def DescomLU(matriz):
    U = np.array(matriz, dtype=float)
    n = len(U)
    L = np.eye(n)  #inicializamos L como la matriz identidad de n*n

    for i in range(n):
        for j in range(i+1, n):
            # alculamos el factor de eliminación
            f = U[j][i] / U[i][i]
            L[j][i] = f
            #actualizamos la fila correspondiente de U
            U[j, i:] = U[j, i:] - f * U[i, i:]

    return L, U

#generamos una matriz aleatoria de entradas enteras de nxn para probar la función
n = 4
matriz = np.random.randint(1, 10, size=(n, n))
print("Matriz original:")
print(matriz)
print("\nDescomposición LU:")
U, L = DescomLU(matriz)
print("Matriz U:")
print(U)
print("\nMatriz L:")
print(L)
```

d) Convierte tu programa en una funcion y usala para crear un programa completo para resolver el sistema (ec. 1), realizando una doble sustitucion inversa (back-sustitution).

```python
import numpy as np

def solucion(L, U, b):
    y = np.zeros(len(L), dtype=float) #inicializamos el vector solución
    for i in range(len(L)):
        y[i] = (b[i] - np.dot(L[i, :i], y[:i])) / L[i][i] #sustitución hacia adelante para resolver L*y = b

    x = np.zeros(len(U), dtype=float) #inicializamos el vector solución
    for i in range(len(U)-1, -1, -1):
        x[i] = (y[i] - np.dot(U[i, i+1:], x[i+1:])) / U[i][i] #sustitución hacia atrás para resolver U*x = y
        
    return x, y

M = [[2, 1, 4, 1],
     [3, 4, -1, -1],
     [1, -4, 1, 5],
     [2, -2, 1, 3]]

b = [-4, 3, 9, 7]

L,U= DescomLU(M)

x, y = solucion(L, U, b)

print("Vector solución x (para A*x = b):")
print(x)
print("\nVector intermedio y (para L*y = b):")
print(y)
```

Escribe un programa para evaluar tu expresion para  Hmn con m y n arbitrarios; cuando la partıcula en el pozo es un electron, el pozo tiene un ancho 5A y  a =10 eV. (La masa y la carga de un electron son 9.1094× 10−31 kg y 1.6022 × 10−19 C respectivamente.)

```python
import numpy as np

#declaramos variables globales
L = 5e-10 #m
a = 10.0 * 1.6022e-19 #J
M = 9.1094e-31               
hbarra = 1.054571817e-34 

#definimos la función que calculará las entradas Hmn

def Hmn(m,n):
    if m!=n and ((m%2==0 and n%2==0) or (m%2!=0 and n%2!=0)):
        return 0
    elif m==n:
        return ((hbarra**2)*(np.pi**2)*(n**2))/(2*M*(L**2))+a/2
    elif m!=n and ((m%2==0 and n%2!=0) or (m%2!=0 and n%2==0)):
        return -(8*m*n*a)/((np.pi**2)*(m**2-n**2)**2)
```

c) En teoría, la matriz H es infinitamente grande, por lo que no podemos calcular 
todos sus valores propios. Pero podemos obtener una solución bastante precisa

Modifica el programa que escribiste para el inciso anterior para crear una matriz

matriz usando la función apropiada de la biblioteca numpy.linalg y entonces imprime 
(en unidades de electron-volts) los primeros diez niveles de energía del pozo cuántico, 
dentro de esta aproximación. 

Deberías encontrar, por ejemplo, que la energía del estado fundamental del sistema es de 
alrededor de 5,84 eV. (Hint: ten en cuenta que los índices matriciales en Python comienzan 
en cero, mientras que los índices en expresiones algebraicas estándar, como las anteriores, 
comienzan en uno. Ten en cuenta esto en tu programa).

```python
#ahora creamos una matriz de 10x10 cuyas entradas serán Hmn
def crear_matriz_H(N): #con N la dimensión de la matriz
    H = np.zeros((N, N))
    for i in range(N):
        for j in range(N):
            H[i][j] = Hmn(i+1, j+1)
    return H

H_10x10=crear_matriz_H(10) #creamos la matriz de 10x10 y la mostramos
H_10x10
```

```python
#ahora sacamos sus eigenvalores usando np.linalg
q=1.6022e-19 #C

E=np.linalg.eigvals(H_10x10) #sacamos los eigenvalores de la matriz H_10x10 en eV
print("Primeros 10 niveles de energía (eV):")

print(np.sort(E/q)) #ordenamos los eigenvalores de menor a mayor
```

Presta mucha atención a la normalización de la función de onda: debe satisfacer la 
condición:

∫₀ᴸ |ψ(x)|² dx = 1.

¿Esto es cierto para tu función de onda?

```python
import numpy as np
import matplotlib.pyplot as plt

#no podemos sumar hasta infinito, así que tomaremos un número N de términos
#escogemos el tamaño del pozo como L=5e-10 m
def E(x, ψ_0, L=5e-10, N=10):
    densidad = 0
    for n in range(1, N+1):
        densidad += ψ_0[n-1] * np.sin((np.pi * n * x) / L)
    return abs(densidad)**2

#usaremos cuadratura gaussiana para integrar
def gauss(f, a, b, ψ_0, N=100):
    nodes, weights = np.polynomial.legendre.leggauss(N)
    nodes = 0.5 * (nodes + 1) * (b - a) + a
    weights *= 0.5 * (b - a)
    integral = sum(w * f(node, ψ_0) for node, w in zip(nodes, weights))
    return integral

valores, vectores = np.linalg.eig(H)

ψ_0 = vectores[:, np.argmin(valores)]  #vector propio correspondiente al menor eigenvalor
#es decir el estado fundamental

a, b = 0, 5e-10
N_gauss = 100  #número de puntos para la cuadratura gaussiana
integral_gauss = gauss(E, a, b, ψ_0)

print(f"El valor de la integral mediante cuadratura gaussiana es: {integral_gauss}")

X = np.linspace(a, b, 1000)
F_val = np.array([E(x, ψ_0) for x in X])  # Use list comprehension to evaluate E(x, ψ_0) for each x
F_val /= np.max(F_val)  # Normalize

plt.title("Densidad de probabilidad")
plt.plot(X, F_val, "-")
plt.grid()
plt.xlabel("x (metros)")
plt.ylabel("|ψ(x)|²")
plt.show()

#X=np.linspace(0,5e-10,1000)
#E_1p=np.array([E_1(x) for x in X])
#E_1p/=np.max(E_1p)
```
